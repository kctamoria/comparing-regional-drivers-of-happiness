# Research Question

__What drivers (factors that are linked to happiness, taken from WorldData.info) of happiness (as measured by the World Happiness Report) most heavily influence the happiness levels within Asian and non-Asian countries and how do these factors differ?__

## Background and Prior Work

Happiness is an intrinsic goal for individuals and government leaders, becoming a priority for the countries' researchers and policymakers. Due to this focus, the World Happiness Report (WHR) was created in 2012 to rank countries on their "national happiness", utilizing a national survey to do so.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) These rankings provide insight into the general well-being of a nation, shaped and contextualized by socioeconomoic and political factors. This project uses contextual data from WorldData.info in conjunction with the WHR to investigate the drivers of happiness across Asian countries as opposed to the rest of the world.

Previous works have explored this idea in multiple countries. For example, in China, education was found to have a significant impact on the happiness (and also economic) levels of the citizens. Furthermore, the level of education that an individual received had a consistent relationship with the level of happiness that they reported: higher education equated to stronger happiness scores.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Another study explored this question among a broad group of countries, particularly in the context of socio-economic factors, where higher-wage workers were found to have a higher subjective well-being score than the rest of the population, which was consistent in every country studied.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) Moreover, a study among 166 developed/developing countries from 2005 to 2020 found that increased income generally had a positive effect on happiness, but in developing countries, this relationship forms an inverted U-shape, where excessive income resulted in a negative impact on happiness levels <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2).

# Hypothesis



We predict that the drivers of happiness in Asian countries will significantly differ from the rest of the world, with certain factors such as education, economic stability, and social support having a greater impact. Previous studies have revealed that in countries like China, education has a strong influence on happiness, and that in developing countries, the link between income and happiness follows a non-linear pattern. These findings support the idea that cultural and economic contexts shape how happiness is driven and experienced in different parts of the world.

# Data

The ideal dataset would include a comprehensive, multi-year report of the happiness levels of most of the countries worldwide, along with some contextual (potential) drivers for the result. The data should be collected annually over the past 5 to 10 years, which would allow for cross-sectional and longitudinal comparisons. The variables would include the happiness scores from the World Happiness Report and the main socio-economic, health, and cultural indicators. These data would be organized by country and year to analyze how these factors that influence happiness level differ between Asian and non-Asian countries.

Potential real data sets include the previous years World Happiness Report datasets as well as cross cultural psychology studies on how cultural differences play a role in subjective well-being.

# Ethics & Privacy

- With the way our research question is phrased (happiness as our main variable), there could be a lot of variability and subjectivity within our measures. There is a risk of imposing Western definitions of “well-being” onto countries with drastically different cultural factors. This can be an ethical concern when comparing cultures.
    - To address it, we utilized a dataset that includes a wide range of “happiness drivers” to adequately compare nations fairly, acknowledging cultural differences, but giving us many variables to compare our findings.
- With large survey datasets, there may be an underrepresentation of developing countries while overrepresenting high-income countries. Furthermore, response bias is something that may be difficult to address if citizens in a specific country underreport their happiness due to social standards, as opposed to another country overreporting. There may also be a bias when reaching out to rural communities in these countries, as surveys are typically conducted electronically.
    - To address this issue, we will make note of our data’s limitations, and make sure not to make generalizations about Asia or specific countries as a whole.
- Our findings may be misused for political purposes, where stereotypes can be made about different cultures that may be found to be less “happy” or unable to achieve happiness.
    - To address this, our data will not be used to rank countries, rather, it will be used to acknowledge the differences in what makes a country happy in Asia, compared to the rest of the world.
